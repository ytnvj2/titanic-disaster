```python
import pandas as pd,numpy as np,os

raw_data_path=os.path.join(os.path.pardir,'data','raw')
train_data_path=os.path.join(raw_data_path,'train.csv')
test_data_path=os.path.join(raw_data_path,'test.csv')

train_df=pd.read_csv(train_data_path,index_col='PassengerId')
test_df=pd.read_csv(test_data_path,index_col='PassengerId')
```

```python
train_df.info()
```

```python
test_df.info()
```

```python
test_df['Survived']=-888
```

```python
df=pd.concat([train_df,test_df],sort=False)
```

```python
df.info()
```

```python
df.columns
```

```python
df=df[[ 'Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket',
       'Fare', 'Cabin', 'Embarked','Survived']]
```

```python
df.head()
```

# Univariate Analysis

```python
cat_cols=['Pclass','Name','Sex','Ticket','Cabin','Embarked','Survived']
num_cols=['Age','Fare','SibSp','Parch']
```

Let's construct Histograms for the Numerical Variables

```python
import matplotlib.pyplot as plt
for i in num_cols:
    df[i].plot(kind='hist')
    plt.xlabel(i)
    plt.show()
```

```python
for i in num_cols:
    df[i].plot(kind='box')
    plt.show()
```

```python
for i in cat_cols:
    df[i].value_counts().plot(kind='bar')
    plt.xlabel(i)
    plt.show()
```

# Bivariate Analysis

```python
import seaborn as sns
for i in num_cols:
    sns.boxplot(x=df['Survived'],y=df[i])
    plt.show()
```

```python
for i in ['Pclass','Sex','Embarked']:
    pd.crosstab(df['Survived'],df[i]).plot(kind='bar')
    plt.show()
```
